```python
# Import necessary libraries
import os
import pandas as pd
import pyodbc
```

## Merge all the settlement files into one 

```python
# Vertically concat all the original csv settlement files into one 

# Directory containing your CSV files
input_directory = 'C:/Users/petrd/Desktop/JetBrains/settlements/'
output_file = 'merged_settlements.csv'

# Create an empty DataFrame with specified column names to store the merged data
column_names = ['COMPANY_ACCOUNT', 'MERCHANT_ACCOUNT', 'BATCH_NUMBER', 'PSP_REFERENCE', 'ORDER_REF', 'PAYMENT_METHOD', 'DATE', 'TIMEZONE', 'TYPE', 'MODIFICATION_REFERENCE', 'CURRENCY', 'NET', 'FEE', 'GROSS']

merged_data = pd.DataFrame(columns=column_names)

# Loop through all CSV files in the directory
for filename in os.listdir(input_directory):
    if filename.endswith('.csv') and filename != 'merged_settlements.csv':
        file_path = os.path.join(input_directory, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, sep=';')
        
        # Append the data to the merged_data DataFrame starting from the second row
        merged_data = pd.concat([merged_data, df], ignore_index=True)


# Save the merged data to a new CSV file
merged_data.to_csv(output_file, index=False, sep=';')

print(f'Merged data saved to {output_file}')
```

## Row count check

```python
# Storing all the original files into one dictionary for further checks

# Directory containing your CSV files
input_directory = 'C:/Users/petrd/Desktop/JetBrains/settlements/'

# Initialize an empty dictionary to store DataFrames
dataframes = {}

# Loop through all CSV files in the directory
for filename in os.listdir(input_directory):
    if filename.endswith('.csv') and filename != 'merged_settlements.csv':
        file_path = os.path.join(input_directory, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, sep=';')
        
        # Store the DataFrame in the dictionary with the file name as the key
        dataframes[filename] = df

# Access each DataFrame by its file name
for filename, df in dataframes.items():
    print(f"Data from '{filename}':")
    print(df.head())
```

```python
# Getting number of rows accross all original files within the directory
# This will help check if all rows were loaded into merged_settlements.csv

# Initiate variables to store the counts 
total_row_count = 0
csv_file_count = 0

# Iterate through the DataFrames in the 'dataframes' dictionary and count total row count
for df_name, df in dataframes.items():
    
    # Count the number of rows in the current DataFrame
    num_rows = len(df)
    
    # Add the file row count to the total row count
    total_row_count += num_rows

# Count number of csv files within directory except merged_settlements.csv
for filename in os.listdir(input_directory):
    if filename.endswith('.csv') and filename != 'merged_settlements.csv':
        csv_file_count += 1
        
print(f"Total number of rows across all files: {total_row_count}")
```

```python
# Check if total row count is same as total row count of merged data - if yes, all data are loaded
if total_row_count == len(merged_data):
    print(f"Count rows in merged_settlements.csv ({len(merged_data)}) corresponds with total row count across all original csv files ({total_row_count}).")
else:
    print("Row counts DOES NOT MATCH.")
```
